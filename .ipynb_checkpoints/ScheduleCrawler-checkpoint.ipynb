```python
import re
import time
import requests
from bs4 import BeautifulSoup
```

```python
class Department:
    def __init__(self,name,code):
        self.name = name
        self.code = code
```

### Read name and code of each department.

```python
departmentsList = open('departments.txt','r')
departments = []
for department in departmentsList:
    t = re.split('~',department)
    departments.append(Department(t[0],t[1]))
```

### Read URLs of each department.

```python
with open('urls.txt') as f:
    urlsList = [line.rstrip() for line in f]
# Line breaks at the end of each line created a problem in using dictionary
urls = []
for url in urlsList:
    t = re.split('~',str(url))
    urls.append((t[0],t[1]))
```

```python
currentTerm = 'donem=2019/2020-2'

schoolURL = 'https://registration.boun.edu.tr/scripts/sch.asp?donem='
```

```python
a = schoolURL+currentTerm+urls[12][1]
print(a)
page = requests.get(a)
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```
